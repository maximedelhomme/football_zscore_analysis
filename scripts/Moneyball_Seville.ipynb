# Cleaning the dataset

```python
def cleaning_num_col(df):
    df_cleaned = df.copy()
    for col in df_cleaned.columns:
        if df_cleaned[col].dtype == 'object':
            df_cleaned[col] = df_cleaned[col].astype(str).str.replace('%', '', regex=False).str.strip().str.replace(',', '.', regex=False)
            try:
                df_cleaned[col] = df_cleaned[col].astype(float)
            except ValueError:
                pass
    return df_cleaned

df_clean = cleaning_num_col(df)


df_clean.info()
```

# Le système de scoring

```python
from sklearn.preprocessing import StandardScaler
```

Utilisation de sklearn pour normaliser les scores sur 100 après le calcul des différents scores.

```python
def evaluer_roles_par_zscore_poste(df, kpi_dict):
    df_result = df.copy()

    for role, contenu in kpi_dict.items():
        print(f"🔍 Évaluation en cours pour le rôle : {role}")
        sous_scores = contenu['sous_scores']

        # Sélection des joueurs du rôle
        role_df = df_result[df_result["role_eval"] == role].copy()
        if role_df.empty:
            print(f"⚠️ Aucun joueur pour le rôle {role}")
            continue

        # Collecte des KPI à utiliser
        all_features = list(set(
            feat for ss in sous_scores.values()
            for feat in ss["features"].keys()
        ))

        # Extraire les colonnes
        role_stats = role_df[all_features].copy()

        # Standardisation (Z-score)
        scaler = StandardScaler()
        role_stats_scaled = pd.DataFrame(
            scaler.fit_transform(role_stats),
            columns=[f"{feat}_z_{role}" for feat in all_features],
            index=role_stats.index
        )

        # Percentiles (entre 0 et 1)
        role_stats_percentiles = pd.DataFrame(index=role_stats.index)
        for feat in all_features:
            role_stats_percentiles[f"{feat}_pct_{role}"] = role_stats[feat].rank(pct=True)

        # Ajout des Z-scores et percentiles au df principal
        df_result = df_result.join(role_stats_scaled, how="left")
        df_result = df_result.join(role_stats_percentiles, how="left")

        # Calcul du score global pondéré
        total_score = pd.Series(0, index=role_stats.index, dtype=float)
        for ss in sous_scores.values():
            poids_ss = ss["poids"]
            ss_score = sum(
                role_stats_scaled[f"{feat}_z_{role}"] * poids_feat
                for feat, poids_feat in ss["features"].items()
                if f"{feat}_z_{role}" in role_stats_scaled.columns
            )
            total_score += ss_score * poids_ss

        # Normalisation sur 100
        mean = total_score.mean()
        std = total_score.std()
        score_100 = (50 + (total_score - mean) / std * 10).clip(0, 100)

        # Stocker le score global
        df_result.loc[role_stats.index, f'score_{role}_100'] = score_100

        print(f"✅ {len(role_stats)} joueurs évalués pour le rôle {role}")

    return df_result
```

Création d'un nouveau fichier avec les variables d'identification (nom, poste, club ...) avec leurs scores associés.

```python
# Lancer l'évaluation (si pas encore fait)
df_evalue = evaluer_roles_par_zscore_poste(df_clean, kpi_dict)

# Définir les colonnes de base (informations joueur)
colonnes_utiles = [
    "Nom", "Club", "Âge", "Taille", "Division", "Pied", "M. poste", "role_eval"
]

# Colonnes de scores globaux normalisés sur 100
colonnes_scores = [col for col in df_evalue.columns if col.startswith("score_") and col.endswith("_100")]

# Colonnes des Z-scores
colonnes_zscores = [col for col in df_evalue.columns if "_z_" in col]

# Colonnes des percentiles
colonnes_percentiles = [col for col in df_evalue.columns if "_pct_" in col]

# Créer le DataFrame final
df_export = df_evalue[colonnes_utiles + colonnes_scores + colonnes_zscores + colonnes_percentiles]

# Exporter vers un fichier CSV
df_export.to_csv("scores_joueurs_detail.csv", index=False)

# Vérification rapide
df_export.head()
```
